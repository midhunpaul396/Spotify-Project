# Creating Pandas DataFrame from list_cursor object

```python
spotifyDataset = panda.DataFrame(list_cursor)
spotifyDataset
```

# Chart 1 Top 10 tracks after 2000 with highest popularity

```python
dataFrameNew = spotifyDataset[spotifyDataset['Year'] >= 2000]
dataFrameNew = dataFrameNew.sort_values(by='Popularity', ascending=False).head(10)
dataFrameNew
```

# Chart 2 Top 10 Tracks of Eminem

```python

dataFrameEminem = spotifyDataset[spotifyDataset['Artist'] == 'Eminem']
dataFrameEminem = dataFrameEminem.sort_values(by='Popularity', ascending=False).head(10)
dataFrameEminem
```

# Chart 3 Top 10 Genre with highest popularity average

```python
# group the DataFrame by 'Genre' column and calculate mean of 'Popularity' for each group
genre_popularity = spotifyDataset.groupby('Genre')['Popularity'].mean()

# sort the resulting Series in descending order and select top 10 rows
top10_popularity_genres = genre_popularity.sort_values(ascending=False).head(10)

top10_popularity_genres_df = top10_popularity_genres.to_frame()

top10_popularity_genres_df
```

# Chart 4 Artists with highest average song releases per year

```python
# Filter the dataset to include only the top 10 artists based on the number of songs released
top10_artists = spotifyDataset.groupby('Artist')['Title'].count().sort_values(ascending=False).head(10).index.tolist()
dataFrameTwo = spotifyDataset[spotifyDataset['Artist'].isin(top10_artists)]

# Group the filtered DataFrame by 'Artist' and 'Year' columns and count the number of songs for each group
grouped = dataFrameTwo.groupby(['Artist', 'Year'])['Title'].count().reset_index()

# Calculate the average number of songs released by each artist in a year
avg_songs_per_artist = grouped.groupby('Artist')['Title'].mean().sort_values(ascending=False).head(10)

# Convert the resulting Series to a DataFrame and reset the index
avg_songs_per_artist_df = avg_songs_per_artist.to_frame().reset_index()

# Rename the columns to reflect the data
avg_songs_per_artist_df.columns = ['Artist', 'Average number of songs released']

# Display the resulting DataFrame
avg_songs_per_artist_df
```

# Using Flask to Generate Visualizations

```python
from flask import Flask, render_template
import pandas as pd
import json

app = Flask(__name__)

@app.route('/')
def index():
    return render_template('index.html')

@app.route('/chart1')
def chart1():
    data = [['Title', 'Popularity']]
    for _, row in dataFrameNew.iterrows():
        data.append([row['Title'], row['Popularity']])
    chart_data = json.dumps(data)
    chart_title = 'Top 10 tracks after 2000 with the highest popularity'
    h_title = 'Song'
    v_title = 'Popularity'
    return render_template('visualizations.html', chart_data=chart_data, chart_title=chart_title, h_title=h_title, v_title=v_title)

@app.route('/chart2')
def chart2():
    data = [['Genre', 'Popularity']]
    for genre, row in top10_popularity_genres_df.iterrows():
        data.append([genre, row['Popularity']])
    chart_data = json.dumps(data)
    chart_title = 'Top 10 Genre with the highest popularity average'
    h_title = 'Genre'
    v_title = 'Popularity'
    return render_template('visualizations.html', chart_data=chart_data, chart_title=chart_title, h_title=h_title, v_title=v_title)

@app.route('/chart3')
def chart3():
    data = [['Title', 'Popularity']]
    for _, row in dataFrameEminem.iterrows():
        data.append([row['Title'], row['Popularity']])
    chart_data = json.dumps(data)
    chart_title = 'Top 10 tracks of Eminem'
    h_title = 'Song'
    v_title = 'Popularity'
    return render_template('visualizations.html', chart_data=chart_data, chart_title=chart_title, h_title=h_title, v_title=v_title)

@app.route('/chart4')
def chart4():
    data = [['Artist', 'Average number of songs released']]
    for _, row in avg_songs_per_artist_df.iterrows():
        data.append([row['Artist'], row['Average number of songs released']])
    chart_data = json.dumps(data)
    chart_title = 'Average number of songs released'
    h_title = 'Artist'
    v_title = 'Average number of songs released'
    return render_template('visualizations.html', chart_data=chart_data, chart_title=chart_title, h_title=h_title, v_title=v_title)

if __name__ == '__main__':
    app.run()
```
